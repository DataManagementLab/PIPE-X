```python
# Base path (usually the home directory of the project)
BASE_PATH = '../..'

MODELS = [  # models able to handle NaN
    'DT',
    'GBT',
    'HGB'
]

EXPERIMENT_NAME = 'sc_imputer_nonredundant'
EXPERIMENT_DATA_MIN = 'adult_noisy_0.000000'
EXPERIMENT_DATA_MAX = 'adult_noisy_0.800000'

# Experiment configuration to use (relative to base path)
EXPERIMENT_CONFIG = f'experiments/yaml_files/exp_{EXPERIMENT_NAME}.yaml'

# Folder containing experiment results (relative to base path)
RESULT_PATH = 'experiments'

# Default sample size
DEFAULT_SAMPLE_SIZE = 100

# Disable displaying detail tables
SHOW_DETAIL_TABLES = False
```

Note
- Feature weights for Leave Out Impact for Destroyer step across different Destroyer probabilities don't change because

- Feature weights for One-by-one Impact for any step before Destroyer in pipeline, such as Replace Marital Status,
  across different Destroyer probabilities don't change because Destroyer is not called as it is later in the pipeline.
